```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
%matplotlib inline
```

```python
df=pd.read_csv(r"C:\Users\OWNER\Desktop\VICTOR\python\py-master\ML\1_linear_reg\Exercise\canada_per_capita_income.csv")
df
```

```python
plt.scatter(df.year,df["per capita income (US$)"],color="red",marker="+")
plt.xlabel("Year")
plt.ylabel("Per capita income")
```

```python
x=df.drop("per capita income (US$)",axis="columns")
x
```

```python
y=df["per capita income (US$)"]
y
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[2020]])
```

```python
import pickle
```

```python
with open("Regressio with pickle","wb") as file:
    pickle.dump(reg,file)
```

```python
with open("Regressio with pickle","rb") as file:
    mp=pickle.load(file)
```

```python
mp.predict([[2020]])
```
