```python
import pandas as pd
df=pd.read_csv(r"C:\Users\OWNER\Desktop\VICTOR\python\py-master\ML\9_decision_tree\Exercise\titanic.csv")
df
```

```python
inputs=df[["Pclass","Sex","Age","Fare"]]
target=df["Survived"]
```

```python
inputs
```

```python
inputs.isnull().sum()
```

```python
mean_age=inputs["Age"].mean()
```

```python
mean_age
```

```python
inputs["Age"]=inputs["Age"].fillna(mean_age)
```

```python
inputs["Age"].isnull().sum()
```

```python
target.isnull().sum()
```

```python
inputs["Pclass"].unique()
```

```python
dummies=pd.get_dummies(inputs["Sex"],prefix="sex")
```

```python
dummies
```

```python
merged=pd.concat([inputs,dummies],axis="columns")
merged
```

```python
inputs=merged.drop("Sex",axis="columns")
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```
